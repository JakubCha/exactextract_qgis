```python
# Function to calculate sum_per_label and unique_label_desc
def calculate_sums(row):
    label_id = np.array(row['values']).flatten()
    fraction_values = np.array(row['coverage']).flatten()
    
    # Create a dictionary to store the sum for each unique label_id
    sum_per_label = {}
    
    # Iterate through unique label_id values
    unique_labels = np.unique(label_id)
    for label in unique_labels:
        # Filter fraction_values based on the current label_id
        values_for_label = fraction_values[label_id == label]
        
        # Calculate the sum and store it in the dictionary
        sum_per_label[label] = np.sum(values_for_label)
    
    # Sort sum_per_label and create a string with unique labels sorted by descending order
    sorted_sum_per_label = {k: v for k, v in sorted(sum_per_label.items(), key=lambda item: item[1], reverse=True)}
    unique_label_desc = ', '.join(map(str, sorted_sum_per_label.keys()))
    
    # Update the row with the calculated values
    row['sum_per_label'] = sorted_sum_per_label
    row['unique_label_desc'] = unique_label_desc
    
    return row
```

```python
%%timeit
def calculate_sums_vectorized(row):
    label_id = pd.Series(row['values'])
    fraction_values = pd.Series(row['coverage'])
    
    # Calculate the sum for each unique label_id using vectorized operations
    sum_per_label = fraction_values.groupby(label_id).sum().sort_values(ascending=False)
    
    # Create a string of the unique labels sorted by descending order
    unique_label_desc = ', '.join(map(str, sum_per_label.index))
    
    # Update the row with the calculated values
    row['sum_per_label'] = sum_per_label.to_dict()
    row['unique_label_desc'] = unique_label_desc
    
    return row

rresults_new = result.iloc[:1000].apply(calculate_sums_vectorized, axis=1)
```

```python
from joblib import Parallel, delayed
from tqdm import tqdm

# Define the number of cores to use for parallel processing
n_jobs = 16

# Apply the function to each row using parallel processing
df_new = Parallel(n_jobs=n_jobs)(delayed(calculate_sums)(row) for _, row in tqdm(result.iterrows()))

# Convert the list of rows back to a DataFrame
df_new = pd.DataFrame(df_new)
```
